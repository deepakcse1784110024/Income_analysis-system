```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\dell\Desktop\canada_per_capita_income.csv")
%matplotlib inline
plt.xlabel("df.year",fontsize=20)
plt.ylabel("df['per capita income (US$)']",fontsize=20)
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
rg=linear_model.LinearRegression()
rg.fit(df[['year']],df[['per capita income (US$)']])
rg.predict(2020)
rg.predict(1999)
rg.coef_
rg.intercept_
%matplotlib inline
plt.xlabel('year')
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df[['per capita income (US$)']])
plt.plot(df.year,rg.predict(df[['year']]),color='red')
```
